```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(df['electric_production'],lags=40)
plot_pacf(df['electric_production'],lags=40)
plt.show
```

```python
train_size = 300

df_train = df.iloc[:train_size]
df_test = df.iloc[train_size:]
```

```python
df_train = df_train.asfreq('MS')
from statsmodels.tsa.arima.model import ARIMA
ARIMA_model = ARIMA(df_train['electric_production'], order=(2,1,4)).fit(method_kwargs={"warn_convergence": False})
print(ARIMA_model.aic)
```

```python
ARIMA_model.summary()
```

```python
# Predict the electric production
df_pred = ARIMA_model.predict(start=300, end=396)

# Plot actual vs. predicted
fig = plt.figure(figsize=(20, 10))
plt.title('ARIMA Predictions', fontsize=20)
plt.plot(df_test['electric_production'], label='actual', color='blue')
plt.plot(df_pred, label='predicted', color='red')
plt.legend(fontsize =20, loc='upper left')
```

```python
rmse = sqrt(mean_squared_error(df_test['electric_production'], df_pred))
print("RMSE:", round(rmse,2))
```

```python
diff_season = diff1 - diff1.shift(12)
diff_season = diff_season.dropna(inplace=False)
```

```python
diff_season.plot(figsize=(10,4))
plt.title('Electric Production Over Time', fontsize=20)
plt.ylabel('Electric Production', fontsize=16)
plot_acf(diff_season,lags=40)
plot_pacf(diff_season,lags=40)
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
SARIMA_model = SARIMAX(df_train['electric_production'],order=(4,1,6),seasonal_order=(2, 1, 3, 12)).fit(method='powell')
```

```python
df_pred = SARIMA_model.predict(start=300, end=396)

# Plot actual vs. predicted
fig = plt.figure(figsize=(20, 10))
plt.title('SARIMA Predictions', fontsize=20)
plt.plot(df_test['electric_production'], label='actual', color='blue')
plt.plot(df_pred, label='predicted', color='red')
plt.legend(fontsize =20, loc='upper left')
```

```python
rmse = sqrt(mean_squared_error(df_test['electric_production'], df_pred))
print("RMSE:", round(rmse,2))
```

```python
# Calculate residuals
residuals = df_test['electric_production'] - df_pred

# Plot residuals
plt.figure(figsize=(10, 5))
plt.plot(residuals)
plt.title('Residuals of SARIMA Model')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()

# Normality test of residuals
from scipy.stats import normaltest
stat, p = normaltest(residuals.dropna())  # Drop NA for handling missing data if any
print('Statistics=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('The residuals seem to be normally distributed (fail to reject H0)')
else:
    print('The residuals do not seem to be normally distributed (reject H0)')

# ACF plot of residuals
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(residuals.dropna(), lags=40)
plt.show()
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

results = acorr_ljungbox(residuals, lags=[10], return_df=True)

print('Ljung-Box test statistic:', results['lb_stat'].iloc[0])
print('Ljung-Box p-value:', results['lb_pvalue'].iloc[0])

lb_stat = results['lb_stat'].iloc[0]
lb_pvalue = results['lb_pvalue'].iloc[0]

if lb_pvalue > 0.05:
    print('The residuals do not have significant autocorrelation (fail to reject H0)')
else:
    print('The residuals have significant autocorrelation (reject H0)')
```

```python
import numpy as np
from scipy.stats import shapiro, ttest_1samp
from statsmodels.stats.diagnostic import acorr_ljungbox, het_breuschpagan
from statsmodels.api import OLS, add_constant

# Test for zero mean
mean_test = ttest_1samp(residuals, 0)
print("Mean Test p-value:", mean_test.pvalue)

# Test for constant variance
ols_results = OLS(residuals, add_constant(np.arange(len(residuals)))).fit()
bp_test = het_breuschpagan(ols_results.resid, ols_results.model.exog)
print("Breusch-Pagan Test p-value:", bp_test[3])

```
